```python
from sklearn.cluster import KMeans
from sklearn import datasets
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import metrics
```

```python
iris = datasets.load_iris()
```

```python
x = pd.DataFrame(iris.data, columns = ['Sepal Lenght', 'Sepal Width', 'Petal Lenght','Petal Width'])
```

```python
plt.scatter(x['Petal Lenght'], x['Petal Width'], c = 'red')
```

```python
model = KMeans(n_clusters = 3, max_iter = len(x))
model.fit(x)
y_kmeans = model.predict(x)
```

```python
y_kmeans
```

```python
accuracy = metrics.adjusted_rand_score(iris.target, y_kmeans)
accuracy
```

```python
plt.scatter(x['Petal Lenght'], x['Petal Width'], c = y_kmeans)
```
