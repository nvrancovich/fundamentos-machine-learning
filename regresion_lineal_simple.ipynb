```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('salarios_32849f3e-7f53-4139-b334-b8de8beb144a.csv')
```

```python
x = df.iloc[:,:-1].values
y = df.iloc[:,1].values
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
regressor = LinearRegression()
regressor.fit(X_train, Y_train)
```

###### Regresión en comparación a los datos que utilizo

```python
viz_train = plt
viz_train.scatter(X_train, Y_train, color = 'blue')
viz_train.plot(X_train, regressor.predict(X_train), color = 'red')
viz_train.show()
```

###### Regresión en comparación a los datos que no utilizo (asignados para testeo)

```python
viz_train = plt
viz_train.scatter(X_test, Y_test, color = 'blue')
viz_train.plot(X_train, regressor.predict(X_train), color = 'red')
viz_train.show()
```
