```python
from sklearn.cluster import KMeans
from sklearn import datasets
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import metrics
from sklearn.preprocessing import StandardScaler
import numpy as np
```

```python
wines = datasets.load_wine()
scaler = StandardScaler()
scaler.fit(wines.data)
x_scaled = scaler.transform(wines.data)
x = pd.DataFrame(x_scaled,columns=np.array(wines.feature_names))
```

```python
plt.scatter(x.alcohol, x.magnesium)
```

```python
model = KMeans(n_clusters = 3, max_iter = len(x))
model.fit(x)
y_kmeans = model.predict(x)
accuracy = metrics.adjusted_rand_score(wines.target, y_kmeans)
accuracy
```

```python
plt.scatter(x.alcohol, x.magnesium, c = y_kmeans)
```
